```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ddrop = [9,8,90,85,80,70,70,65,55,60,70,65,55]
railgun = [75,70,60,65,60,45,55,50,40,40,35,35,30]
x_axis = np.arange(0,130,10)
```

```python
ddrop, = plt.plot(x_axis,ddrop,marker='o',color='b',label='Danger Drop')
railgun, = plt.plot(x_axis,railgun,marker='s',color='g',label='RailGun')
plt.xlim(0,120)
plt.ylim(0,95)
plt.grid()
plt.title('Roller Coaster Speeds')
plt.xlabel('Time (seconds)')
plt.ylabel('Speed (MPH)')
plt.legend(handles=[ddrop,railgun],loc='best')
plt.show()
```
