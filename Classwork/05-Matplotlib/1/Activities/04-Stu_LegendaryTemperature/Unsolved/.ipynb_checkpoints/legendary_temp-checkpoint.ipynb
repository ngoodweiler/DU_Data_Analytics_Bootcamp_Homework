```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temps = [39,42,51,62,72,82,86,84,77,65,55,44]
temps
```

```python
temps_c = [(x-32)*5/9 for x in temps]
temps_c
```

```python
x_axis = np.arange(1,13,1)
temps, = plt.plot(x_axis,temps,label='Fahrenheit',marker='s',color='b')
temps_c = plt.plot(x_axis,temps_c,label='Celsius',marker='+',color='r')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.show()
```
