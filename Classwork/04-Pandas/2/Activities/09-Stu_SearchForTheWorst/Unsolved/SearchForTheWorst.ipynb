```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Create reference to CSV file
csv_path = "Resources/Soccer2018Data.csv"

# Import the CSV into a pandas DataFrame
soccer_2018_df = pd.read_csv(csv_path, low_memory=False)
soccer_2018_df
```

```python
# Collect a list of all the unique values in "Preferred Position"
positions_cleaned = soccer_2018_df.dropna(how='any')
positions_cleaned['Preferred Position'].value_counts()

```

```python
positions_cleaned['Preferred Position'].nunique()
```

```python
# Looking only at strikers (ST) to start
st_df=positions_cleaned.loc[positions_cleaned['Preferred Position'] == 'ST']
st_df
```

```python
# Sort the DataFrame by the values in the "ST" column to find the worst
st_sorted= st_df.sort_values('ST')
st_sorted
# Reset the index so that the index is now based on the sorting locations
new_index = st_sorted.reset_index(drop=True)
new_index.head()
```

```python
# Save all of the information collected on the worst striker
worst_player = new_index.loc[0,:]
worst_player.to_csv("worst_striker.csv",index=True)
```
