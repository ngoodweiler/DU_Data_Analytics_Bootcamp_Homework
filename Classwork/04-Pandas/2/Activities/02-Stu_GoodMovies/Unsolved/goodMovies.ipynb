```python
# Dependencies
import pandas as pd
import csv
```

```python
# Load in File from resources
# 'movie_scores.csv'
movies_csv= ("Resources/movie_scores.csv")
```

```python
# Read and display with pandas
movies_df = pd.read_csv(movies_csv)
movies_df.head()
```

```python
# List all the columns the table provides
movies_columns = movies_df.columns
movies_columns
```

```python
# only care about Imdb, so create a new table that takes the Film and all the columns relating to IMDB
movies_by_name = movies_df.set_index("FILM")
movies_by_name.loc[:,["IMDB","IMDB_norm","IMDB_norm_round","IMDB_user_vote_count"]]
```

```python
# List only movies whose ratings are over 7 (out of 10) in IMDB
imdb_greater_7 = movies_df.loc[movies_df['IMDB'] > 7,:]
# this will give all
#print(imdb_greater_7)
imdb_greater_7.head()
```

```python
# Find lesser-known movies to watch, with fewer than 20K votes
less_known = movies_df.loc[movies_df["IMDB_user_vote_count"] < 20000,:]
less_known.head()
```

```python
# Finally, export this file to an Excel spreadsheet -- without the DataFrame index.
less_known.to_excel("whateveryouwant.xlsx",index=False)
```
